```python
import sys
import subprocess
import time
import os
import pickle
from IPython.display import Image, display
import glob
import matplotlib.pyplot as plt
```

```python
start_time = time.time()
```

```python
result = subprocess.run(
    [sys.executable, "parallel_runner.py"],
    capture_output=True,  # Captures stdout and stderr
    text=True             # Returns strings instead of bytes
)

print("STDOUT:\n", result.stdout)
print("STDERR:\n", result.stderr)
print("Return Code:", result.returncode)
```

```python
result_files = sorted(glob.glob("wcETEL_results/dataset_*_weights.pkl"))

plt.figure(figsize=(10, 6))

for file in result_files:
    with open(file, 'rb') as f:
        result = pickle.load(f)
        masses1 = result['masses1']
        plt.plot(masses1, alpha=0.3)

N = 300
plt.axhline(1/N, color='red', linestyle='--', label='1/300')

plt.title("Overlay of Actual Weights for 50 Clean Datasets")
plt.xlabel("Index")
plt.ylabel("Probability Weight")
plt.legend()
plt.show()
```

```python
end_time = time.time()
print(f"\nTotal runtime for all datasets (sequential): {end_time - start_time:.2f} seconds")
```
